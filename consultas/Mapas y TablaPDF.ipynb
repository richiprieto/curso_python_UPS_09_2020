# Mapas

```python
# pip3 install ipyleaflet
# pip3 install --user pyroutelib3
from ipyleaflet import Map, AntPath
from pyroutelib3 import Router # Import the router
```

```python
# Ruta en bruto
m = Map(center=(-2.1524996,-80.0501411), zoom=12)

ant_path = AntPath(
    locations=[
        [-2.155587,-79.9053877], 
        [-2.1771473,-79.8955064], 
        [-2.1989223,-79.8994984], 
        [-2.1939236,-79.9313442],
    ],
    dash_array=[1, 10],
    delay=1000,
    color='#7590ba',
    pulse_color='#3f6fba'
)
m.add_layer(ant_path)
```

```python
m
```

```python
# Modo de transporte: car, cycle, foot, horse, tram, train
router = Router("car") # Initialise it

coordenadas = ant_path.locations
ruta_completa = []
for coor1,coor2 in zip(coordenadas[0::1], coordenadas[1::1]):

    start = router.findNode(coor1[0], coor1[1]) # Nodos de inicio y fin
    end = router.findNode(coor2[0], coor2[1])
    print(coor1,coor2)
    status, route = router.doRoute(start, end) # Crear la ruta en base a las direcciones proporcionadas en OSM
    if status == 'success':
        routeLatLons = list(map(router.nodeLatLon, route)) # Obtener todas las coordenadas de rutas
        ruta_completa.extend(routeLatLons) # agregar todas las rutas en una sola lista
```

```python
ruta_completa
```

```python
m = Map(center=(-2.1524996,-80.0501411), zoom=12)

ant_path1 = AntPath(
    locations=ruta_completa,
    dash_array=[1, 10],
    delay=2000,
    color='#7590ba',
    pulse_color='#3f6fba'
)
m.add_layer(ant_path1)
```
