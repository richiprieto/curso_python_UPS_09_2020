## Leyendo archivos PDF y Word
### PDF

```python
# pip3 install --user PyPDF2
import PyPDF2
```

```python
# Abrir pdf modo lectura
pdf_dir = open("examples/documento.pdf", "rb")
```

```python
# Leyendo archivo PDF
archivo_pdf = PyPDF2.PdfFileReader(pdf_dir)
```

```python
# Número de paginas
archivo_pdf.numPages
```

```python
#extrayendo texto de un archivo PDF
archivo_pdf.getPage(0).extractText()
```

```python
# extrayendo de multiples páginas
for pagina in range(archivo_pdf.numPages):
    print(archivo_pdf.getPage(pagina).extractText())
```

```python
# Leemos un archivo pdf en modo lectura
pdf_dir2 = open("examples/documento.pdf", "rb")
archivo_pdf2 = PyPDF2.PdfFileReader(pdf_dir)
```

```python
# Creamos PDF 
crear_pdf = PyPDF2.PdfFileWriter()
```

```python
# Lectura de paginas para ir haciendo un append
for pagina in range(archivo_pdf.numPages):
    pagina_append = archivo_pdf.getPage(pagina)
    crear_pdf.addPage(pagina_append)
```

```python
# Lecturas de paginas para ir haciendo un append
for pagina in range(archivo_pdf2.numPages):
    pagina_append = archivo_pdf2.getPage(pagina)
    crear_pdf.addPage(pagina_append)
```

```python
# Se crea un documento pdf en modo lectura
combinado = open("pdfcombinado.pdf", "wb")
```

```python
# creamos pdf
crear_pdf.write(combinado)
```

```python
# Cerrando PDF
combinado.close()
pdf_dir.close()
pdf_dir2.close()
```

### Word

```python
#pip3 install --user python-docx
import docx
```

```python
# Accediendo al documento
documento = docx.Document("examples/documento.docx")
```

```python
# Objeto documento
documento
```

```python
# presentando los parrafos
documento.paragraphs
```

```python
# extrayendo texto de parrafos
documento.paragraphs[0].text
```

```python
# Extrayendo toda la información de parrafos
for parrafo in documento.paragraphs:
    print(parrafo.text)
```

```python
# escribiendo archivos Word
# Agregamos un objeto documento
nuevo_docu = docx.Document()
```

```python
# Agregamos una cabecera
nuevo_docu.add_heading("Hola esto es una cabecera")
```

```python
# Agregamos un parrafo
nuevo_docu.add_paragraph("Esto es un parrafo")
```

```python
# Guardamos el archivo
nuevo_docu.save("nuevo_documento.docx")
```
